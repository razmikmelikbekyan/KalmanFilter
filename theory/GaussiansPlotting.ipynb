```python
import matplotlib.pyplot as plt
from mpl_toolkits.axisartist import SubplotZero
import numpy as np
import scipy.stats as stats
```

```python
def generate_figure():
    
    fig = plt.figure(1, figsize=(10, 6))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)
    
    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)

    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)
    
    return fig, ax
```

## For Presentation

#### Plot 1

```python
fig, ax = generate_figure()

mu = 5
sigma = 1
dist = stats.norm(mu, sigma)

x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
y = dist.pdf(x)

sigma_y = y[25]
sigma_x = dist.ppf(y[25] / 2)
sigma_dx = mu - sigma_x 
sigma_dy = 0

ax.set_xlim(0, 10)
ax.set_ylim(0, 0.6)

ax.set_xticks([])
ax.set_yticks([])

ax.plot(x, y)
ax.vlines(mu, 0, stats.norm.pdf(mu, mu, sigma), color='r')
ax.arrow(mu, sigma_y, -sigma_dx+0.15, sigma_dy, 
         head_width=0.01, head_length=0.15)

ax.annotate('$\sigma_{x_{0}}$', (mu - sigma_dx / 2, y[25] - 0.03), xycoords='data',  fontsize=20)
ax.annotate('$\hat{x}_{0}$', (5.1, 0.02), xycoords='data', fontsize=20)
ax.annotate('$x$', (9.8, 0.02), xycoords='data', fontsize=20)
ax.annotate('$PDF - f(x)$', (0.2, 0.57), xycoords='data', fontsize=20)

plt.tight_layout()
# plt.savefig('gaussian_1')
plt.show()
```

#### Plot 4

```python
fig, ax = generate_figure()

mu_1 = 11
sigma_1 = 2.
dist = stats.norm(mu_1, sigma_1)

x = np.linspace(mu_1 - 3 * sigma_1, mu_1 + 3 * sigma_1, 100)
y = dist.pdf(x)

ax.plot(x, y, linestyle='--', alpha=0.5)
ax.vlines(mu_1, 0, stats.norm.pdf(mu_1, mu_1, sigma_1), color='black', linestyle='--', alpha=0.5)

ax.annotate('$\hat{x}_{1|0}$', (mu_1 -1, 0.02), xycoords='data', fontsize=15)
ax.annotate('$x$', (24.5, 0.02), xycoords='data', fontsize=20)
ax.annotate('$PDF - f(x_1 \;| \; x_{1|0}, \; z_{1})$', (0.2, 0.57), xycoords='data', fontsize=20)

mu_2 = 19
sigma_2 = 1.3
dist = stats.norm(mu_2, sigma_2)

x = np.linspace(mu_2 - 3 * sigma_2, mu_2 + 3 * sigma_2, 100)
y = dist.pdf(x)

ax.plot(x, y, linestyle='--', alpha=0.5)
ax.vlines(mu_2, 0, stats.norm.pdf(mu_2, mu_2, sigma_2),  color='black', linestyle='--', alpha=0.5)

ax.annotate('$z_{1}$', (mu_2 + 0.3, 0.02), xycoords='data', fontsize=15)


k = sigma_1 ** 2 / (sigma_1 ** 2 + sigma_2 ** 2)
mu = mu_1 + k * (mu_2 - mu_1)
sigma = np.sqrt((sigma_1 ** 2  * sigma_2 ** 2 )/ (sigma_1 ** 2 + sigma_2 ** 2))
dist = stats.norm(mu, sigma)

x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
y = dist.pdf(x)

sigma_y = y[35]
sigma_x = dist.ppf(y[35] / 2)
sigma_dx = mu - sigma_x 
sigma_dy = 0

ax.plot(x, y)
ax.vlines(mu, 0, stats.norm.pdf(mu, mu, sigma), color='r')
ax.arrow(mu, sigma_y, -sigma_dx+0.45, sigma_dy, 
         head_width=0.01, head_length=0.15)

ax.annotate('$\sigma_{x_{1|1}}$', (mu - sigma_dx / 2 - 0.4, y[35] - 0.03), 
            xycoords='data',  fontsize=14, fontweight='bold')
ax.annotate('$\hat{x}_{1|1}$', (mu - 1.3, 0.04), xycoords='data', fontsize=15, fontweight='bold')

ax.set_xlim(0, 25)
ax.set_ylim(0, 0.6)

ax.set_xticks([])
ax.set_yticks([])

plt.tight_layout()

# plt.savefig('gaussian_4')
plt.show()
```
